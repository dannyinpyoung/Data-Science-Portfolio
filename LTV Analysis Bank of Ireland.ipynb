# LTV Regulation Policy Analysis: Bank of Ireland Mortgage Pool and Subprime Lending

__The Data__
The data used is scraped from the Bank of Ireland. It is in the format of a PDF, scraped, and then converted to a CSV for analysis on Excel. 

```python
import PyPDF2 as p2
import re
import pandas as pd

#TODO: Create function to read PDF for table 1, scrape into list
def get_table_2_list (year):
    file = open("{}.pdf".format(year), "rb") 
    PDFfile = p2.PdfFileReader(file)
    page_1 = PDFfile.getPage(0)
    Table_2 = page_1.extractText()
    table_2_list = re.sub(' +', ' ', Table_2.replace('\n', ' ')).split(' ')
    table_2_list.insert(134, ' ')
    return table_2_list

#TODO: Create function that generates DF and exports to excel.

def create_df(year):
    table_2_df = pd.DataFrame({
            '>':[],
            '<=':[],
            'Current Balance (Euros)': [],
            '% of Total': [],
            'No. of Loans': [],
            '% of Total': [],
        })
    row_section = 67
    for i in range(0,12): 
        table_2_df.loc[i] = get_table_2_list(year)[row_section:(row_section + 5)]
        row_section += 6
    return table_2_df

create_df(2019)

#TODO: Optional Functions for Analysis

#PDFfile = open("2019.pdf", "rb")
#pdfread = p2.PdfFileReader(PDFfile)
#table_1 = pdfread.getPage(0)
#print(table_1.extractText())
#print(pdfread.getIsEncrypted())
#print(pdfread.getDocumentInfo())
#print(pdfread.getNumPages())
```
